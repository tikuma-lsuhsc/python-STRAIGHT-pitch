```python
import numpy as np
from matplotlib import pyplot as plt
from math import pi
```

```python
n = 100
dt = 1 / n
x = np.arange(3 * n + 1) * dt

g = np.exp(-pi * x**2) * np.sinc(x) ** 2
dg = np.zeros_like(g)
tf = x != 0
x1 = x[tf]
dg[tf] = (
    2
    * (-pi * x1**3 * np.sinc(x1) + x1 * np.cos(x1) - np.sin(x1))
    * np.exp(-pi * x1**2)
    * np.sinc(x1)
    / x1**2
)

plt.plot(x, g)
plt.plot(x, dg)
```
